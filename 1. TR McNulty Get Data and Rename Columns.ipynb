```python
import pandas as pd
import re

file = './data/anes_columns_all_6.txt'

with open(file) as f:
    text = f.readlines()
raw = [x.strip() for x in text]

i = 0
columns={}
while i <= len(raw)-1:
    c_id = raw[i].strip()
    columns[c_id]=raw[i+2].replace(':','').replace('/','_').replace(' ','_').replace('-','_').replace('___','_')+'_'+c_id
    i+=3

#REMOVE EXTRA UPPERCASE LABEL PORTION    
for k,v in columns.items():
    columns[k] = re.sub(r"^[A-Z_]*_", "", v) 
```

```python
#couldn't read 2016 data due to a unicode-8 character not recognized...
#data_file = '2016_data/anes_timeseries_2016_dta/anes_timeseries_2016.dta'
data_file = 'data/anes_timeseries_cdf.dta'
df = pd.read_stata(data_file)
df.rename(columns=columns,inplace=True)
```
